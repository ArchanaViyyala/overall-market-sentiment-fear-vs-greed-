```python
# Convert Timestamp IST to datetime (with proper format)
trader['Timestamp IST'] = pd.to_datetime(trader['Timestamp IST'], format='%d-%m-%Y %H:%M', errors='coerce')

# Remove rows with missing critical data
trader.dropna(subset=['Execution Price', 'Size USD', 'Side', 'Closed PnL'], inplace=True)

# Standardize text columns
trader['Side'] = trader['Side'].str.lower()
```

```python
# Add profit/loss classification
trader['PnL_type'] = np.where(trader['Closed PnL'] > 0, 'Profit', 'Loss')

# Calculate volume = execution price * size
trader['volume'] = trader['Execution Price'] * trader['Size USD']

# Extract date to merge with sentiment
trader['date'] = trader['Timestamp IST'].dt.date
```

```python
# Convert date
sentiment['date'] = pd.to_datetime(sentiment['date'])
sentiment['classification'] = sentiment['classification'].str.strip().str.title()
```

```python
# Add profit/loss classification
trader['PnL_type'] = np.where(trader['Closed PnL'] > 0, 'Profit', 'Loss')

# Calculate volume = execution price * size
trader['volume'] = trader['Execution Price'] * trader['Size USD']

# Extract date to merge with sentiment
trader['date'] = trader['Timestamp IST'].dt.date
```

```python
# Prepare sentiment data
sentiment['date'] = sentiment['date'].dt.date

# Merge datasets on date
merged = pd.merge(trader, sentiment[['date', 'classification']], on='date', how='left')

# Check final dataframe
print(merged.head())
```

```python
# Check for missing values in the merged key columns
print(merged[['classification', 'pnl_type']].isnull().sum())

# Drop rows with missing classification or pnl_type
merged_clean = merged.dropna(subset=['classification', 'pnl_type'])
```

```python
os.makedirs("outputs", exist_ok=True)
```

```python
#Plot 1: Profit vs Loss by Market Sentiment
plt.figure(figsize=(8, 5))
sns.countplot(data=merged_clean, x='classification', hue='pnl_type')
plt.title('Profit vs Loss by Market Sentiment')
plt.xlabel('Market Sentiment')
plt.ylabel('Number of Trades')
plt.legend(title='PnL Type')
plt.tight_layout()
plt.savefig("outputs/profit_by_sentiment.png")
plt.close()

```

```python
# Standardize columns in both datasets before merging
trader.columns = trader.columns.str.strip().str.lower().str.replace(' ', '_')
sentiment.columns = sentiment.columns.str.strip().str.lower().str.replace(' ', '_')

# Create a date column in both
trader['date'] = trader['timestamp_ist'].dt.date
sentiment['date'] = pd.to_datetime(sentiment['date']).dt.date

# Merge
merged = pd.merge(trader, sentiment[['date', 'classification']], on='date', how='left')

# Check columns
print(merged.columns.tolist())
```

```python
# Plot 2: Volume Distribution by Sentiment
plt.figure(figsize=(8, 5))
sns.boxplot(data=merged_clean, x='classification', y='volume')
plt.title('Trade Volume by Market Sentiment')
plt.xlabel('Market Sentiment')
plt.ylabel('Trade Volume (USD)')
plt.tight_layout()
plt.savefig("outputs/volume_by_sentiment.png")
plt.close()
```

```python
# Plot 3: Trade Side (Buy/Sell) by Sentiment
plt.figure(figsize=(8, 5))
sns.countplot(data=merged_clean, x='classification', hue='side')
plt.title('Trade Side by Market Sentiment')
plt.xlabel('Market Sentiment')
plt.ylabel('Number of Trades')
plt.legend(title='Trade Side')
plt.tight_layout()
plt.savefig("outputs/side_by_sentiment.png")
plt.close()
```
